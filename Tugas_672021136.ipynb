- Pie Chart

```python
import matplotlib.pyplot as plt

nama_permen = ['Mentos', 'Kopiko', 'Golia', 'Yupie', 'Fisherman']
jumlah_permen = [52, 39, 78, 13, 78]
warna = ('#964e00', '#c7c0b9','#ff4f69', '#e8dc35', '#a04cf5')
target_penjualan = (0,0.1,0,0,0)

indeks_kopiko = nama_permen.index('Kopiko')

peluang_kopiko = jumlah_permen[indeks_kopiko] / sum(jumlah_permen) * 100

labels = [nama_permen[i] + f" ({jumlah_permen[i]})" for i in range(len(nama_permen))]

plt.figure(figsize=(8, 6))
plt.pie(jumlah_permen, labels=labels, autopct='%1.0f%%', startangle=140, explode = target_penjualan)
plt.title('Persentase Peluang Memilih Permen Kopiko')

plt.show()
```

Dari diagram pie diatas kita dapat mengetahui bahwa peluang memilih permen kopiko adalah 15%

Kita memiliki penggunaan CPU rata-rata per jam untuk
komputer pekerja selama seminggu. Setiap baris data
mewakili satu hari dalam seminggu yang dimulai dengan
Senin. Setiap kolom data adalah satu jam dalam sehari
dimulai dengan 0 menjadi tengah malam. Buat bagan yang
menunjukkan penggunaan CPU selama seminggu. Anda
harus dapat menjawab pertanyaan-pertanyaan berikut
menggunakan bagan:
1. Jam berapa pekerja biasanya makan siang?
2. Apakah pekerja tersebut bekerja pada akhir pekan?
3. Pada hari apa pekerja mulai bekerja pada komputer mereka
pada malam hari?

- Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

datapenggunaan_cpu = [
    [2, 2, 4, 2, 4, 1, 1, 4, 4, 12, 22, 23, 45, 9, 33, 56, 23, 40, 21, 6, 6, 2, 2, 3], # Senin
    [1, 2, 3, 2, 3, 2, 3, 2, 7, 22, 45, 44, 33, 9, 23, 19, 33, 56, 12, 2, 3, 1, 2, 2], # Selasa
    [2, 3, 1, 2, 4, 4, 2, 2, 1, 2, 5, 31, 54, 7, 6, 34, 68, 34, 49, 6, 6, 2, 2, 3], # Rabu
    [1, 2, 3, 2, 4, 1, 2, 4, 1, 17, 24, 18, 41, 3, 44, 42, 12, 36, 41, 2, 2, 4, 2, 4], # Kamis
    [4, 1, 2, 2, 3, 2, 5, 1, 2, 12, 33, 27, 43, 8, 38, 53, 29, 45, 39, 3, 1, 1, 3, 4], # Jumat
    [2, 3, 1, 2, 2, 5, 2, 8, 4, 2, 3, 1, 5, 1, 2, 3, 2, 6, 1, 2, 2, 1, 4, 3], # Sabtu
    [1, 2, 3, 1, 1, 3, 4, 2, 3, 1, 2, 2, 5, 3, 2, 1, 4, 2, 45, 26, 33, 2, 2, 1], # Minggu
]

hari = ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu']
jam = list(range(1, 25))

plt.figure(figsize=(12, 8))
sns.heatmap(datapenggunaan_cpu, cmap='coolwarm', xticklabels=jam, yticklabels=hari)
plt.title('Heatmap Penggunaan CPU per Hari dan Jam')
plt.xlabel('Jam')
plt.ylabel('Hari')
plt.show()
```

```python
import numpy as np

# Data penggunaan CPU untuk hari Senin hingga Jumat
data_hari_kerja = datapenggunaan_cpu[:5]

# Hitung rata-rata penggunaan CPU pada setiap jam
rata_rata_cpu_per_jam = np.mean(data_hari_kerja, axis=0)

# Temukan jam dengan rata-rata penggunaan CPU terendah
jam_makan_siang = np.argmin(rata_rata_cpu_per_jam) + 1  # Argmin memberikan indeks jam terkecil, ditambah 1 karena indeks dimulai dari 0
print("Jam makan siang biasanya adalah jam:", jam_makan_siang)
```

```python
# Cek apakah ada penggunaan CPU pada Sabtu atau Minggu
bekerja_akhir_pekan = any([sum(data_hari) > 0 for data_hari in datapenggunaan_cpu[-2:]])

if bekerja_akhir_pekan:
    print("Pekerja bekerja pada akhir pekan.")
else:
    print("Pekerja tidak bekerja pada akhir pekan.")
```

```python
data_jam_malam_hari = [hari[18:] for hari in datapenggunaan_cpu]

# Temukan hari dengan penggunaan CPU tertinggi pada malam hari
hari_mulai_kerja_malam = np.argmax([max(data_hari) for data_hari in data_jam_malam_hari])

print("Pekerja mulai bekerja pada komputer pada malam hari pada hari:", hari[hari_mulai_kerja_malam])
```
